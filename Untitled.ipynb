```python
SHAPE = [1080,1920]
def get_filter(frame,bg_sub,brighten=True,blur=False):
    gray=cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    denoise=cv2.GaussianBlur(gray,(71,71),0)
    if blur:
        fg_mask = bg_sub.apply(denoise, None, 0.001)
    elif brighten:
        bright=cv2.convertScaleAbs(gray, alpha=1, beta=50)
        fg_mask = bg_sub.apply(bright, None, 0.001)
    img=cv2.Canny(denoise,10,10)
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (10,10))
    closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
    combined=cv2.bitwise_and(fg_mask,closing)
    return combined

def train_bg_subtractor(cap, bg_sub, num, brighten=True, blur=False):

    for i in range(num):
        ret,frame=cap.read()
        gray=cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        if brighten:
            bright=cv2.convertScaleAbs(gray, alpha=1, beta=50)
            fg_mask=bg_sub.apply(bright, None, 0.001)
        elif blur:
            denoise=cv2.GaussianBlur(gray,(71,71),0)
            fg_mask=bg_sub.apply(denoise, None, 0.001)
        else:
            fg_mask=bg_sub.apply(gray, None, 0.001)
    return bg_sub

def save_frame(frame, file_name, flip=True):
    # flip BGR to RGB
    if flip:
        cv2.imwrite(file_name, np.flip(frame, 2))
    else:
        cv2.imwrite(file_name, frame)
        
        
def process_vid(vid_name,output_vid,brighten=True,blur=False):
    num_frames=count_frames(vid_name)
    cap=cv2.VideoCapture(vid_file)
    train_frames=round(num_frames/4)
    bg_sub=cv2.createBackgroundSubtractorMOG2(history=train_frames,detectShadows=True)
    bg_sub=train_bg_subtractor(cap, bg_sub,train_frames)
    fourcc = cv2.VideoWriter_fourcc(*"MJPG")
    
    writer = cv2.VideoWriter(output_vid, fourcc, cap.get(cv2.CAP_PROP_FPS), (SHAPE[1], SHAPE[0]), True)
    for i in range(num_frames-train_frames):
        ret,frame=cap.read()
        mask=get_filter(frame,bg_sub)
        matches=get_contours(mask)
        img1=[]
        img1=draw_boxes(frame,matches)
        writer.write(img1)
    writer.release()
    cap.release()
        
        

process_vid(vid_file,'trial2.avi')
```
