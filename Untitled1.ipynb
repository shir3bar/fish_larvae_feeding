```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
vid_file='11.12.2019 - frames 64935-65415.avi'
#'03.06.20 13DPH.mp4'
cap=cv2.VideoCapture(vid_file)
```

```python
ret, frame1=cap.read()
```

```python
plt.imshow(frame1)
```

```python
ret, frame2=cap.read()
```

```python
plt.imshow(frame2)
```

```python
grayA = cv2.cvtColor(frame1, cv2.COLOR_BGR2GRAY)
grayB = cv2.cvtColor(frame2, cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize=(16,16))
plt.imshow(cv2.absdiff(grayB,grayA))
plt.show()
```

```python

diff_image = cv2.absdiff(grayB, grayA)

# perform image thresholding
ret, thresh = cv2.threshold(diff_image, 20, 255, cv2.THRESH_BINARY)

# plot image after thresholding
plt.imshow(thresh, cmap = 'gray')
```
